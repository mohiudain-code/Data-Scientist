```python
import pandas as pd
df=pd.read_csv('train_update.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(df['SalePrice'])
plt.show()
```

```python
q1=df['SalePrice'].quantile(0.25)
q3=df['SalePrice'].quantile(0.75)
```

```python
iqr=q3-q1
min_r= q1-(1.5*iqr)
max_r= q3+(1.5*iqr)
max_r
```

```python
df=(df[df['SalePrice']<=max_r])

sns.distplot(df['SalePrice'])
plt.show()

df.head(5)
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
ft=FunctionTransformer(func=lambda x:x**2)
ft.fit(df[['SalePrice']])
df['SalePrice_FT']=ft.transform(df[['SalePrice']])
df
```

```python
plt.subplot(1,2,1)
sns.distplot(df['SalePrice'])
plt.title('Non-Normal Distribution')
plt.subplot(1,2,2)
sns.distplot(df['SalePrice_FT'])
plt.title('Normal Distribution')
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
ft=FunctionTransformer(func=np.log1p)
ft.fit(df[['SalePrice']])
df['SalePrice_FT']=ft.transform(df[['SalePrice']])
```

```python
plt.subplot(1,2,1)
sns.distplot(df['SalePrice'])
plt.title('Non-Normal Distribution')
plt.subplot(1,2,2)
sns.distplot(df['SalePrice_FT'])
plt.title('Normal Distribution')
plt.show()
```
