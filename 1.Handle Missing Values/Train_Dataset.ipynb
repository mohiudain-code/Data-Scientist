6.Scikitlearn->Missing values Handle Filling

```python
import pandas as pd
df=pd.read_csv('train_update.csv')
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='most_frequent')
df=pd.DataFrame(si.fit_transform(df.select_dtypes(include='object')))
df.head()
df.isnull().sum()
```

7.One Hot Encoding->Catogrical to Numerical

```python
from sklearn.preprocessing import OneHotEncoder
OHT=OneHotEncoder(drop='first')
import pandas as pd
df=pd.read_csv('train_update.csv')
data=df['Street']
var=pd.get_dummies(data)
arr=OHT.fit_transform(var).toarray()
pd.DataFrame(arr)
```
