```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import FunctionTransformer
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df.head(3)
```

```python
sns.scatterplot(x="Age",y="EstimatedSalary",hue="Purchased",data=df)
plt.show()
```

# Check Normalization yes or not

```python
sns.kdeplot(df['Age'])
plt.show()
```

# Normalization Better

```python
ft=FunctionTransformer(lambda x:x**2)
ft.fit(df['Age'])
df['Age']=ft.transform(df[['Age']])
sns.kdeplot(df['Age'])
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df['Purchased']
```

# Guassion NB

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

# MultiNomial NB

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

# Bernolli NB

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
