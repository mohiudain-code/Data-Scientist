OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
arr=[['Normal', 'Abnorml', 'Partial', 'AdjLand', 'Alloca', 'Family']]
Oe=OrdinalEncoder(categories=arr)
df['SaleCondition_new']=Oe.fit_transform(df[['SaleCondition']])
df
```

```python
import pandas as pd
df=pd.DataFrame({'Size':['s','m','l','xl','m','s','l','m','xl']})
df
df['Size'].unique()
Oe=OrdinalEncoder(categories=[['s', 'm', 'l', 'xl']])
df['Size_new']=Oe.fit_transform(df[['Size']])
df
```

Map Function Through Ordinal Encoding

```python
import pandas as pd
df=pd.read_csv('train_update.csv')
df['SaleCondition'].unique()
dict={
    'Normal':0,
    'Abnorml':1,
    'Partial':2,
    'AdjLand':3,
    'Alloca':4,
    'Family':5,
}
df['SaleCondition']=df['SaleCondition'].map(dict)
df.head(20)
```

```python
import pandas as pd
df=pd.DataFrame({'Area':['5 Marla','10 Marla','1 Kanal','2 Kanal']})
dict={
    '5 Marla':5,
    '10 Marla':10,
    '1 Kanal':1,
    '2 Kanal':2
}
df['Area_new']=df['Area'].map(dict)
df
```
